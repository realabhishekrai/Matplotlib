```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
days = np.arange(1,16)
temperature = np.random.randint(32,36,15)
```

```python
temp_df = pd.DataFrame({"days":days, "temperature":temperature})
sns.lineplot(x="days",y="temperature", data=temp_df)
plt.show()
```

# Working on real data set

```python
tips_df = sns.load_dataset("tips")
tips_df.head()
```

```python
tips_df.shape
```

```python
sns.lineplot(x="total_bill", y ="tip", data = tips_df)
```

```python
sns.lineplot(x="tip", y = "total_bill", data=tips_df)
```

```python
# now check for tip and size
sns.lineplot(x = "tip", y = "size", data = tips_df)
```

```python
# now check for size and total bill
sns.lineplot(x='size', y='total_bill', data=tips_df)
```

```python
# more parameters
sns.lineplot(x='size', y='total_bill', data=tips_df, hue='sex', style = 'sex',\
            palette = 'hot', dashes = False, marker = "d", legend = 'full')
plt.title("Line Plot", fontsize=20)
plt.xlabel("size",fontsize=15)
plt.ylabel("total bill",fontsize=15)
plt.show()
```

```python
# sns.set for background
plt.figure(figsize=(10,6))
sns.set(style="darkgrid")
sns.lineplot(x='size', y='total_bill', data=tips_df, hue='sex', style = 'sex',\
            palette = 'hot', dashes = False, marker = "d", legend = 'full')
# you can also give list to marker for more than one lines
plt.title("Line Plot", fontsize=20)
plt.xlabel("size",fontsize=15)
plt.ylabel("total bill",fontsize=15)
plt.show()
```
