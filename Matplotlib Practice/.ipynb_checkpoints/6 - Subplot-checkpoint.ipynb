```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1,2,1)
plt.subplot(1,2,2)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.pie([1])
plt.subplot(2,2,2)
plt.pie([5,7])
plt.subplot(2,2,3)
plt.pie([56,4,4,3,3,42])
plt.subplot(2,2,4)
plt.pie([5,6,43,2,34,2])
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(3,2,1)
plt.subplot(3,2,2)
plt.subplot(3,2,3)
plt.subplot(3,2,4)
plt.subplot(3,2,5)
plt.subplot(3,2,6)
plt.show()
```

```python
import numpy as np

np.random.seed(42)

plt.figure(figsize= (10,6))

x = list(np.random.randint(1,10,25))
y = list(np.random.randint(5,10,25))

plt.subplot(2,2,1)
plt.plot(x)

plt.subplot(2,2,2)
plt.hist(x)          # frequecny count

plt.subplot(2,2,3)
plt.bar(x,y)

plt.subplot(2,2,4)
plt.pie(y)

plt.show()
```
