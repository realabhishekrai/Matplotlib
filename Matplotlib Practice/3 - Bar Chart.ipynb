```python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import style
```

```python
classes = ['Python','R','AI','ML','DS']
class1_students = [30,10,20,25,10] # out of 100 students in each class
class2_students = [40,5,20,20,20]
class3_students = [35,5,20,15,15]
```

```python
plt.bar(classes,class1_students, width = 0.4, align = 'edge', color = 'y', edgecolor = "m",\
       linewidth = 4, alpha = 0.9, linestyle = '--', label = 'Class 1 students', visible = True) 
# align can be 'center' which is default and 'edge' is optional
# edge color
# linewidth
# alpha = for transparency of graph
# visible = True Can will visible but when visible = False graph will not visible
```

```python
plt.barh(classes,class1_students) # for horizontal plotting of data
```

# Part 2 Working with multiple data

```python
style.use('ggplot')
plt.figure(figsize=(10,6))
plt.bar(classes,class1_students, width = 0.4, align = 'edge', color = 'k', edgecolor = "m",\
       linewidth = 4, alpha = 0.9, linestyle = '--', label = 'Class 1 students', visible = True) 
plt.title('Classed of different topics')
plt.xlabel('Classes', color = 'r')
plt.ylabel('Number of students', color = 'r')
```

# That's how our graph with 2 data

```python
plt.bar(classes,class1_students)
plt.bar(classes,class2_students)
```

# Now arranging both data

```python
class_index = np.arange(len(classes))
plt.bar(class_index,class1_students, width=0.4, label = 'class 1')
plt.bar(class_index+0.4,class2_students, width = 0.4, label = 'class 2')
plt.legend()
```
