```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
classes = ['Python','R','Machine Learning','Artificial Intelligence','Data Science']
class1_students = [45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students, labels = classes)
plt.show()
```

```python
explode = [0.1,0,0.2,0,0] # use for slicing 
plt.pie(class1_students, labels = classes, explode = explode)
plt.show()
```

```python
color = ['r','g','y','b','m']
plt.pie(class1_students,labels = classes, c= color)
```
