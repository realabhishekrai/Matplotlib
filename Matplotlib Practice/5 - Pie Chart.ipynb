```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
classes = ['Python','R','Machine Learning','Artificial Intelligence','Data Science']
class1_students = [45,15,35,25,30]
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students, labels = classes)
plt.show()
```

```python
explode = [0.1,0,0.2,0,0] # use for slicing 
plt.pie(class1_students, labels = classes, explode = explode)
plt.show()
```

```python
colors = ['r','b','y','c','m']
explode = [0,0,0.1,0,0.2]
plt.pie(class1_students,labels = classes, colors = colors, explode = explode, autopct = "%0.2f%%", \
       shadow = True, radius = 1.2)
plt.show()
```

# startangle for starting of graph from particular angle

```python
colors = ['r','b','y','c','m']
explode = [0,0,0.1,0,0.2]
plt.pie(class1_students,labels = classes, colors = colors, explode = explode, autopct = "%0.2f%%", \
       shadow = True, radius = 1.2, startangle = 90)
plt.show()
```

# textprops for text manupulation

```python
colors = ['r','b','y','c','m']
explode = [0,0,0.1,0,0.2]
textprops = {"fontsize":14}
plt.pie(class1_students,labels = classes, colors = colors, explode = explode, autopct = "%0.2f%%", \
       shadow = True, radius = 1.2, startangle = 90, textprops = textprops)
plt.show()
```

# Now checking for some extra parameter

```python
wedgeprops = {'linewidth':2, 'width':1, "edgecolor":'k'}
plt.pie(class1_students, labels = classes, radius = 1.6, autopct = "%0.2f%%", textprops = {'fontsize':14}, labeldistance = 1.2,\
       wedgeprops = wedgeprops, shadow = True, rotatelabels = True, explode = explode)
plt.show()
```
