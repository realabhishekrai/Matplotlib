```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_students_age = np.random.randint(18,45,100)
py_students_age = np.random.randint(15,40,100)
```

```python
plt.hist(ml_students_age)
plt.title('Machine Learning Students Age Histogram')
plt.xlabel('Age')
plt.ylabel('Number of students')
```

```python
bins = [15,20,25,30,35,40,45] # use bins to give range to graph
plt.hist(ml_students_age, bins, rwidth = 0.8) # rwidth use for range width 1 for 100% (we used 0.8 that means 80%)
plt.xlabel('Age')
plt.ylabel('Number of students')
plt.title('Machine Learning Students Age Histogram')
```

# Orientation for histogram

```python
bins = [15,20,25,30,35,40,45] 
plt.hist(ml_students_age, bins, rwidth = 0.8, orientation = 'horizontal', color = 'm', label = 'ML Students') # orientation horizontal or vertical
plt.xlabel('Age')
plt.ylabel('Number of students')
plt.title('Machine Learning Students Age Histogram')
plt.legend()
```

# One histogram with two representation of data

```python
bins = [15,20,25,30,35,40,45] 
plt.figure(figsize = (10,6))  # for figure size
plt.hist(ml_students_age, bins, rwidth = 0.5, label = 'ML Studetns')
plt.hist(py_students_age, bins, rwidth = 0.5, label = 'Py Studetns')
plt.title('ML and Python Students Age')
plt.xlabel('Age')
plt.ylabel('Number of students')
plt.legend()
```

# Other and good way to represent multiple data

```python
plt.figure(figsize = (10,6))
plt.hist([ml_students_age,py_students_age],bins, label = ['ML students','Py students'], color =['c','r']\
        , orientation = 'vertical')
plt.xlabel('Age',fontsize = 20)
plt.ylabel('Number of studetns', fontsize=20)
plt.legend()
```

# Adding style module

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize = (10,6))
plt.hist([ml_students_age,py_students_age],bins, label = ['ML students','Py students'], color =['c','r']\
        , orientation = 'vertical')
plt.xlabel('Age',fontsize = 20)
plt.ylabel('Number of studetns', fontsize=20)
plt.legend()
plt.grid()
```
