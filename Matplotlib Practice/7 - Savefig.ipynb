```python
import pandas as pd
import numpy as np
import random
```

```python
ipl = pd.read_csv('C:\\Users\\reala\\Downloads\\matches.csv')
```

```python
winner = ipl['winner']
```

```python
winner.fillna('Nan', inplace=True)
```

```python
win_count = list(winner.value_counts().sort_index())
```

```python
print(win_count)
```

```python
teams = list(pd.unique(winner))
teams = sorted(teams)
```

# Using location to save graph

```python
from matplotlib import style
style.use("ggplot")
plt.figure(figsize=(12,7))
plt.bar(teams,win_count)
plt.title('Match wins IPL')
plt.xlabel('No of match wins')
plt.ylabel('Teams')
plt.xticks(rotation = 75, fontsize=10)
plt.savefig('C:\\Users\\reala\\Downloads\\IPL_Wins',dpi = 100, quality = 99)
```
