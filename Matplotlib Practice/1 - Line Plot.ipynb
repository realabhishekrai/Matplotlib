```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = list(np.arange(1,32))
```

```python
temperature = list(np.random.randint(25,48,31))
```

```python
plt.plot(days,temperature, color = 'b', marker ='o', linestyle = ':' ,linewidth = 2, markersize = 10 )  # shift+tab button for help
plt.title('temperatrue of delhi')
plt.xlabel('days')
plt.ylabel('temperature')
plt.axis([0,31,0,50])  # axis range
plt.show()
```

```python
# shortcut to color marker and linestyle
plt.plot(days,temperature,'bo--')  #color= 'b',marker = 'o',linestyle = '--' Note:- you can only give marker and linestyle also
plt.title('delhi temperature', fontsize = 20)
plt.xlabel('days', fontsize = 20)
plt.ylabel('temperature', fontsize=15)
plt.legend('tem line', loc = 1) # loc = can be 1,2,3,... up to 10
# you can use label in 
```

```python
# # color list 
# b : blue.
# g : green.
# r : red.
# c : cyan.
# m : magenta.
# y : yellow.
# k : black.
# w : white


# **Markers**

# =============    ===============================
# character        description
# =============    ===============================
# ``'.'``          point marker
# ``','``          pixel marker
# ``'o'``          circle marker
# ``'v'``          triangle_down marker
# ``'^'``          triangle_up marker
# ``'<'``          triangle_left marker
# ``'>'``          triangle_right marker
# ``'1'``          tri_down marker
# ``'2'``          tri_up marker
# ``'3'``          tri_left marker
# ``'4'``          tri_right marker
# ``'s'``          square marker
# ``'p'``          pentagon marker
# ``'*'``          star marker
# ``'h'``          hexagon1 marker
# ``'H'``          hexagon2 marker
# ``'+'``          plus marker
# ``'x'``          x marker
# ``'D'``          diamond marker
# ``'d'``          thin_diamond marker
# ``'|'``          vline marker
# ``'_'``          hline marker
# =============    ===============================

# **Line Styles**

# =============    ===============================
# character        description
# =============    ===============================
# ``'-'``          solid line style
# ``'--'``         dashed line style
# ``'-.'``         dash-dot line style
# ``':'``          dotted line style
# =============    ===============================
```

```python
# Using grid 
plt.plot(days,temperature,'bo--')  #color= 'b',marker = 'o',linestyle = '--' Note:- you can only give marker and linestyle also
plt.title('delhi temperature', fontsize = 20)
plt.xlabel('days', fontsize = 20)
plt.ylabel('temperature', fontsize=15)
plt.legend('tem line', loc = 1) # loc = can be 1,2,3,... up to 10
plt.grid(color='c',linestyle='-', linewidth = 2)  # backgroung  grid(color='r', linestyle='-', linewidth=2)
# you can use label in 
```

```python
# using style method
from matplotlib import style
style.use('ggplot')
plt.plot(days,temperature,'bo--')  #color= 'b',marker = 'o',linestyle = '--' Note:- you can only give marker and linestyle also
plt.title('delhi temperature', fontsize = 20)
plt.xlabel('days', fontsize = 20)
plt.ylabel('temperature', fontsize=15)
plt.legend('tem line', loc = 1) # loc = can be 1,2,3,... up to 10
# you can use label in 
```

```python
# plotting multigraph
delhi_temp = [32,34,28,29,41,37,29,43,36,34,38,32,29,40,35,32]
mumbai_temp = [29,40,35,32,32,34,38,32,29,43,36,28,32,41,37,39]
```

```python
plt.plot(delhi_temp, label='delhi_temp',marker = 'o', linewidth=4, markersize = 12)
plt.plot(mumbai_temp, label = 'mumbai_temp',marker = 'd',linewidth=4,markersize = 12)
plt.axis([0,16,15,50])
plt.title('Delhi & Mumbai Temperature')
plt.xlabel('Days', fontsize = 20)
plt.ylabel('Temperature',fontsize=20)
plt.legend()
plt.grid(color='w',linestyle='--', linewidth = '2')
```
